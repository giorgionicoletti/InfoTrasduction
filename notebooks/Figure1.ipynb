```python
def load_from_mathematica(filename):
    db = pd.read_csv(filename, header = None)
    db = db.map(lambda x: str(x))
    db = db.map(lambda x: x.strip('{}'))

    db = db.map(lambda x: x.replace('*^', 'e'))

    db = db.map(lambda x: float(x))
    return db.values.T
```

```python
Sdot, Iyeta = load_from_mathematica("../data/mathematica/Fig1_ThetaEta5_Sigma5_(x_Sdotmin)_(y_IyEta).txt")
Sdot = Sdot[1:]
Iyeta = Iyeta[1:]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (6, 6))

ax.plot(Sdot, Iyeta, 'k-', lw = 3)

ax.set_xlabel(r'$\dot{S}_a$', fontsize = 30)
ax.set_ylabel(r'$I_{y\eta}$', fontsize = 30)

ax.set_xlim([-0.1, 2.9])
ax.set_ylim([-0.01, 0.47])

ax.set_xticks([0, 2.8])
ax.set_yticks([0, 0.45])

plt.show()
```
